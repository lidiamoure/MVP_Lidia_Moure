## 1. Valores sobre homosexualidade

### 1.1 O que se observa sobre a tolerância a homosexualidade ao longo das waves? E quanto ao período mais recente (entre as waves 6 e 7)?

```sql
select
  wave_chronology,
  neighbors_homosexuals,
  count(*) * 100.0 / sum(count(*)) over (partition by wave_chronology) as response_percentage
from mvp_wvs.gold.fact_wave_all fa
join mvp_wvs.gold.dim_age da
  on fa.age = da.age
join mvp_wvs.gold.dim_age_range ar
  on da.age_range = ar.age_range
where ar.age_range = '15 - 25 years old'
group by wave_chronology, neighbors_homosexuals
```

```sql
select
  wave_chronology,
  justifiable_homosexuality,
  count(*) * 100.0 / sum(count(*)) over (partition by wave_chronology) as response_percentage
from mvp_wvs.gold.fact_wave_all fa
join mvp_wvs.gold.dim_age da
  on fa.age = da.age
join mvp_wvs.gold.dim_age_range ar
  on da.age_range = ar.age_range
where ar.age_range = '15 - 25 years old'
group by wave_chronology, justifiable_homosexuality
```
