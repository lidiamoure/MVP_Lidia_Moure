Neste notebook, serão unidas todas as 5 tabelas em uma só. Em seguida, serão substutuidos os valores numéricos das instâncias pelo texto correspondente, de acordo com o documento de equivalência disponibilizado no site da WVS.

```python
%python
spark.sql("USE CATALOG mvp_wvs")
spark.sql("USE SCHEMA silver")
```

```python
spark.sql("DROP TABLE IF EXISTS mvp_wvs.silver.wave_all")
```

### União das tabelas Wave 3, Wave 4, Wave 5, Wave 6 e Wave 7 na nova tabela wave_all.

```python
%sql
CREATE OR REPLACE TABLE mvp_wvs.silver.wave_all AS
SELECT * FROM mvp_wvs.silver.wave3
UNION ALL
SELECT * FROM mvp_wvs.silver.wave4
UNION ALL
SELECT * FROM mvp_wvs.silver.wave5
UNION ALL
SELECT * FROM mvp_wvs.silver.wave6
UNION ALL
SELECT * FROM mvp_wvs.silver.wave7;
```

```python
%sql
ALTER TABLE mvp_wvs.silver.wave_all
SET TBLPROPERTIES (
  'delta.columnMapping.mode' = 'name'
)
```

### Coluna "Sex"

```python
%sql
-- Conferência dos valores distintos deste atributo.
select distinct sex  
from mvp_wvs.silver.wave_all
```

```python
%sql
-- Criação da coluna "sex_label" e disposição das instâncias em formato texto, correspondendo aos valores numéricos da coluna "Sex"
ALTER TABLE mvp_wvs.silver.wave_all
ADD COLUMN sex_label STRING;

UPDATE mvp_wvs.silver.wave_all
SET sex_label =
    CASE
        WHEN sex = 1 THEN 'Male'
        WHEN sex = -1 THEN 'Don''t Know'
        WHEN sex = -2 THEN 'No answer'
        WHEN sex = 2 THEN 'Female'
        WHEN sex = 9 THEN 'Not Applicable'
        WHEN sex = -5 THEN 'Missing'
    END;
```

```python
%sql
--Conferência de se a coluna "sex_label" foi criada corretamente
select distinct sex_label
from mvp_wvs.silver.wave_all
```

```python
%sql
--Exclusão da coluna "Sex"
ALTER TABLE mvp_wvs.silver.wave_all
DROP COLUMN sex
```
