### Neste notebook,  serão dispostas no schema silver as tabelas apenas com as colunas úteis para criação do pipeline. Em seguida, as colunas serão renomeadas. 

```sql
%python
spark.sql("USE CATALOG mvp_wvs")
spark.sql("USE SCHEMA silver")
```

```sql
CREATE OR REPLACE TABLE mvp_wvs.silver.wave3 AS --cria tabela da wave 3 com as colunas escolhidas
SELECT
    V9,
    V22,
    V60,
    V61,
    V99,
    V101,
    V103,
    V135,
    V179,  
    V181,
    V182,
    V190,
    V197,
    V198,
    V199,
    V200,
    V201,
    V1,    
    V2,
    V238,
    V214,
    V216
FROM mvp_wvs.bronze.wave3;

```
