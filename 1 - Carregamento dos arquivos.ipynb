### Neste notebook, ser√£o salvos os arquivos csv como tabelas no schema bronze. 

```sql
%python
spark.sql("USE CATALOG mvp_wvs")
spark.sql("USE SCHEMA bronze")
```

```sql

%python
file_path = "/Volumes/mvp_wvs/staging/wvs/Wave3.csv" #salva a tabela da wave 3 no schema bronze

df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("sep", ";") \
    .load(file_path)

display(df.limit(10))

table_name = "Wave3"
df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable(table_name)
```
